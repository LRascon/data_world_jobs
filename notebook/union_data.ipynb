```python
#Importamos nuestras librerias
import pandas as pd
import numpy as np
import re
```

```python
jobs = pd.read_csv('/Henry_Final/data_world_jobs/notebook/jobs.csv', encoding='latin1')
benefits = pd.read_csv('/Henry_Final/data_world_jobs/notebook/benefits.csv', encoding='latin1')

companies = pd.read_csv('/Henry_Final/data_world_jobs/notebook/companies.csv', encoding='latin1')
location = pd.read_csv('/Henry_Final/data_world_jobs/notebook/location.csv', encoding='latin1')

skills = pd.read_csv('/Henry_Final/data_world_jobs/notebook/skills.csv', encoding='latin1')
skills_unique = pd.read_csv('/Henry_Final/data_world_jobs/notebook/skills_unique.csv', encoding='latin1')
```

```python
# Reemplaza los valores vacíos por NaN
skills.replace("", np.nan, inplace=True)
```

```python
# Combina las columnas en una sola columna 'skills', excluyendo los NaN y separando con comas
skills['skills'] = skills.apply(lambda row: ', '.join(row.dropna().astype(str)), axis=1)
```

```python
# Elimina las columnas originales
skills.drop(columns=['skill_1', 'skill_2', 'skill_3', 'skill_4', 'skill_5', 'skill_6', 'skill_7'], inplace=True)
```

```python
#unimos jobs y skills
data_lb = pd.concat([jobs, skills], axis=1)
```

```python
# Eliminar las columnas 'min' y 'max'
data_lb.drop(columns=['min', 'max'], inplace=True)
```

```python
data_lb = pd.concat([data_lb, companies], axis=1)
```

```python
data_lb = pd.concat([data_lb, location], axis=1)
```

```python
data_lb['level_job'] = data_lb['level_job'].replace('No disponible', 'N/A')
```

```python
# Definir la expresión regular para encontrar los trabajos remotos
regex = re.compile(r'(remote)', flags=re.IGNORECASE)

# Crear una nueva columna 'onsite_remote' en el DataFrame, asignando 'onsite' como valor por defecto
data_lb['onsite_remote'] = 'onsite'

# Asignar la categoría correspondiente a los trabajos que sean remotos
data_lb.loc[data_lb['location'].str.contains(regex, na=False), 'onsite_remote'] = 'remote'
```

```python
# Definir la nueva expresión regular
regex = re.compile(r'(scientist|engineer|analyst|ml)', flags=re.IGNORECASE)

# Crear una nueva columna 'dim_JobCategory' en el dataframe JOB, asignando 'others' como valor por defecto.
data_lb['dim_JobCategory'] = 'others'

# Asignar la categoría correspondiente a los títulos que hagan match con la expresión regular
data_lb.loc[data_lb['title_job'].str.contains(regex), 'dim_JobCategory'] = data_lb['title_job'].str.extract(regex, expand=False).str.lower()
```

```python
# Reemplazar '-level' con una cadena vacía en la columna 'level_job'
data_lb['level_job'] = data_lb['level_job'].replace(r'-level', '', regex=True)
```

```python
data_lb = data_lb.rename(columns={"salary_average": "annual_salary_usd",})
```

```python
data_lb['annual_salary_usd'] = data_lb['annual_salary_usd'].replace(0, np.nan)
```

```python
# Reemplazar '-level' con una cadena vacía en la columna 'level_job'
data_lb['level_job'] = data_lb['level_job'].replace(r'-level', '', regex=True)
```

```python
# Definir la nueva expresiÃ³n regular
regex = re.compile(r'(scientist|engineer|analyst|ml)', flags=re.IGNORECASE)

# Crear una nueva columna 'dim_JobCategory' en el dataframe JOB, asignando 'others' como valor por defecto.
data_lb['dim_JobCategory'] = 'others'

# Asignar la categorÃ­a correspondiente a los tÃ­tulos que hagan match con la expresiÃ³n regular
data_lb.loc[data_lb['title_job'].str.contains(regex), 'dim_JobCategory'] = data_lb['title_job'].str.extract(regex, expand=False).str.lower()
```

```python
# Reemplazar '-level' con una cadena vacÃ­a en la columna 'level_job'
data_lb['level_job'] = data_lb['level_job'].replace(r'-level', '', regex=True)
```

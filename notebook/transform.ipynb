## Creamos la Tabla Jobs

```python
#creamos el dataframe para la tabla jobs
jobs = df.iloc[:,[0,2,3,8,14]]
```

```python
jobs.head()
```

```python
#Revisamos nuestros valores faltantes
jobs.isna().sum()
```

```python
#Tranformamos los NaN del level en texto descriptivo
jobs['level_job'].fillna('No disponible', inplace=True)
```

```python
#eliminamos el texto del salary range
# Eliminar "USD" de la columna "salario"
jobs['salary_range'] = jobs['salary_range'].str.replace('USD', '')
jobs['salary_range'] = jobs['salary_range'].str.replace('K', '000')
jobs['salary_range'] = jobs['salary_range'].str.replace('-', '')
jobs['salary_range'] = jobs['salary_range'].str.replace('*', '')
jobs['salary_range'] = jobs['salary_range'].str.replace('+', '')
jobs['salary_range'] = jobs['salary_range'].str.replace('EUR', '')
jobs['salary_range'] = jobs['salary_range'].str.replace('Career', '')
jobs['salary_range'] = jobs['salary_range'].str.replace('development', '')

```

```python
valores_unicos = jobs['salary_range'].unique()
```

```python
valores_unicos
```

```python
pattern = r"[^0-9 ]"  # Patrón para buscar caracteres distintos a números y espacios en blanco
```

```python
jobs['salary_range'] = jobs['salary_range'].fillna('0') # Rellenamos los nulos
```

```python
jobs['salary_range'] = jobs['salary_range'].str.replace(pattern, "")  # Reemplaza los caracteres no deseados por vacío
```

```python
new_cols = jobs['salary_range'].str.split(' ', expand=True)
```

```python
new_cols = new_cols.drop([0,2,4],axis=1)
```

```python
new_cols
```

```python
new_cols.dtypes
```

```python
new_cols[['min', 'max']] = new_cols[['min', 'max']].apply(pd.to_numeric)
```

```python
new_cols['salary_average'] = (new_cols['min'] + new_cols['max']) / 2
```

```python
jobs = pd.concat([jobs, new_cols], axis=1)
```

```python
jobs = jobs.drop('salary_reference', axis=1)
```

```python
jobs[['min','max','salary_average']] = jobs[['min','max','salary_average']] .fillna(0)
```

```python
has_nan = jobs.isnull().any().any()
```

## Creamos la Tabla Skills Unicas

```python
# Seleccionamos las columnas de habilidades y combínamos en una sola columna
skills_unique = skills[["skill_1", "skill_2", "skill_3", "skill_4", "skill_5", "skill_6", "skill_7"]].melt()["value"]
```

```python
# Eliminamos las filas con valores nulos o faltantes
skills_unique = skills_unique.dropna()
```

```python
# Eliminamos los elementos duplicados y creamos una lista con los elementos únicos resultantes
unique_skills = skills_unique.unique().tolist()
```

```python
# Imprimimos la lista de habilidades únicas
unique_skills
```

```python
# Lo convertimos en un dataset
skills_ok = pd.DataFrame(unique_skills, columns=["skills"])
```
